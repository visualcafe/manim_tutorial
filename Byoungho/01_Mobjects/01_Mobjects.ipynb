# 1. shape_matchers

다른 Mobject 들을 강조하거나 표시하기 위한 Mobject 입니다.

## Classes

- `BackgroundRectangle` : 배경 직사각형을 만듭니다.
- `Cross` : X 표시를 만듭니다.
- `SurroundingRectangle` : Mobject를 둘러싼 직사각형을 만듭니다.
- `Underline` : 밑줄을 만듭니다.

```python
# BackgroundRectangle 

%%manim -v WARNING --disable_caching -qh ExBackgroundRectangle

config.media_width = "75%"

class ExBackgroundRectangle(Scene):
    def construct(self):
        circle = Circle().shift(LEFT)
        triangle = Triangle().shift(RIGHT)
        background_rectangle_1 = BackgroundRectangle(circle, color=WHITE, fill_opacity=0.5)
        background_rectangle_2 = BackgroundRectangle(triangle, color=WHITE, fill_opacity=0.5)
        self.add(backgroundRectangle1, backgroundRectangle2)
        self.add(circle, triangle)
```

```python
%%manim -v WARNING --disable_caching -qh ExCross

config.media_width = "75%"

class ExCross(Scene):
    def construct(self):
        circle = Circle(color=WHITE)
        cross = Cross(circle, color=RED)
        self.add(circle, cross)
```

```python
%%manim -v WARNING --disable_caching -qh ExSurroundingRectangle

config.media_width = "75%"

class ExSurroundingRectangle(Scene):
    def construct(self):
        circle = Circle()
        surrounding_rectangle_1 = SurroundingRectangle(circle, color=WHITE)
        self.add(surrounding_rectangle_1, circle)
```

```python
%%manim -v WARNING --disable_caching -qh ExUnderline

config.media_width = "75%"

class ExUnderline(Scene):
    def construct(self):
        circle = Circle()
        under_line = Underline(circle, color=YELLOW)
        self.add(under_line, circle)
```
